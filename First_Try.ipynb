```python
import json
import os
from PIL import Image
#import torch
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
from collections import Counter
#from torchvision import transforms
```

```python
def verzeichnisstruktur_anzeigen(data_dir):
    print(f"Data directory: {data_dir}")
    print(f'Inhalt des Verzeichnisses {data_dir}: {os.listdir(data_dir)}')
    names = ("test_dir", "train_dir")
    namensliste = list()
    for name in names:
        if name == "test_dir":
            path = data_dir / "test"
            directory_name = "Testverzeichnis"
        else:
            path = data_dir / "train"
            directory_name = "Trainverzeichnis"
        print(f"\n{directory_name}: {path}")
        print(f"Inhalt des Verzeichnisses {directory_name}: {os.listdir(path)}")
        namensliste.append(path)
    return namensliste
```

```python
data_dir = Path("C:\\Users\\domin\\Desktop\\al5083")

test_dir, train_dir = verzeichnisstruktur_anzeigen(data_dir)
```

```python
# Funktion zum Laden der JSON-Datei
def load_json(json_path):
    with open(json_path, 'r') as f:
        data = json.load(f)
    # Ausgabe der Anzahl der geladenen Bilddaten
    print(f'Geladene Bilddaten in {json_path}: {len(data)}')
    # Ausgabe der Klassenverteilung
    class_counts = Counter(data.values())
    print(f'Klassenverteilung in {json_path}: {sorted(class_counts.items())}')
    return data, class_counts
```

```python
train_json_path = data_dir / "train" / "train.json"
print('Anzahl der Verzeichnisse im Train-Verzeichnis:', len(os.listdir(train_dir)))
print(f"Train JSON path: {train_json_path}")
json_data_train, class_counts_train = load_json(train_json_path)
print()
test_json_path = data_dir / "test" / "test.json"
print('Anzahl der Verzeichnisse im Test-Verzeichnis:', len(os.listdir(test_dir)))
print(f"Test JSON path: {test_json_path}")
json_data_test, class_counts_test = load_json(test_json_path)
```
